```python
import sys
import tenpy
import numpy as np
from su2_model import corr_function_ss, SU2Model, corr_function_etap_etam
import matplotlib.pyplot as plt
import yaml
```

```python
yaml_file = 'su2_simulation_VBS.yml'
filename = 'dmrg_VBS_Jp_26.4_6.h5'

#get results of the simulation
results = tenpy.tools.hdf5_io.load('results/' + filename)

with open(yaml_file, 'r') as f:
    data = yaml.full_load(f)
```

```python
print(len(results['sweep_stats']['max_trunc_err']))
print(results['sweep_stats']['max_trunc_err'])

print("expectation value:\n Sz:")
print(results['psi'].expectation_value("Sz"))
print("Etap")
print(results['psi'].expectation_value("Etap"))
print("Etam")
print(results['psi'].expectation_value("Etam"))
print("Correlation length:")
print(results['psi'].correlation_length2())

L = results['psi'].L
corr = corr_function_ss(results['psi'], list(range(L)), list(range(L)))
print("<S S>")
print(corr[0])
print("<Etap Etam>")
print(corr_function_etap_etam(results['psi'], list(range(L)), list(range(L)))[0])
```

```python
#create the sites again so we can plot them
#idk a better way to do this rn
model = SU2Model(data['model_params'])
sites = model.lat.mps_sites() # create sites

# plot nearest neighbor correlations
fig, ax = plt.subplots()
model.lat.plot_sites(ax)
model.lat.plot_coupling_correlations(ax, corr, cmap='RdBu_r', value_func=np.real, vmin=-3/4, vmax=1/4)
ax.set_aspect('equal')
plt.savefig(f'results/a_{filename}_corr.png')

# plot energy per iteration
fig2, ax2 = plt.subplots()
energy_arr = results['sweep_stats']['E']
ax2.plot(np.arange(len(energy_arr)), energy_arr)
plt.savefig(f'results/a_{filename}_energy.png')

#plot maximum truncation error per iteration
fig3, ax3 = plt.subplots()
max_trunc_err_arr = results['sweep_stats']['max_trunc_err']
ax3.plot(np.arange(len(max_trunc_err_arr)), max_trunc_err_arr)
plt.savefig(f'results/a_{filename}_trunc_err.png')
```
